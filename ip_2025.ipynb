```python
#find max wicket
df['wicket'] = df['wicket_type'].apply(lambda x: 1 if x in ['caught', 'bowled', 'lbw', 'stumped', 'hit wicket'] else 0)
```

```python
df['wicket_type'].value_counts()
```

```python
wicket=df.groupby('bowler').sum()
wicket=wicket['wicket']
```

```python
wicket=pd.DataFrame(wicket)
wicket=wicket.sort_values(by='wicket',ascending=False)
wicket.head(10)
```

economy rate


```python
#Economy Rate
econ=df.groupby('bowler').sum()
econ=(((econ['runs_of_bat'])+(econ['extras']))*6)/(econ['ball'])
econ=econ.sort_values()
econ.head(10)
```

```python
wicket['econ']=econ
wicket
```

```python
sns.lineplot(data=wicket)
```
